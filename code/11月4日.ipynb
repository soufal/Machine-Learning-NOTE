## 手写数字识别系统      
需要识别的数字时32*32像素大小的黑白图像

### 算法：      
1. 收集数据
2. 准备数据：编写函数classify0(),将图像格式转换为分类器使用的list格式
3. 分析数据：
4. 测试算法：
5. 使用算法

```python
from numpy import *
from os import listdir
import operator
```

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    ##计算距离(欧式距离)
    diffMat = tile(inX, (dataSetSize, 1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndicies = distances.argsort()     #距离排序
    classCount = {}
    ##选择距离最小的k个点
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    #print(classCount)
    sortedClassCount = sorted(classCount.items(),
                             key = operator.itemgetter(1), reverse=True)
    print(sortedClassCount)
    return sortedClassCount[0][0]
```

### 准备数据：将图像转换为测试向量

**将一个$32*32$的二进制图像矩阵转换为$1*1024$的向量**     
打开文件，循环读出文件的前32行，并将头32个字符值存储在NumPy数组中，最后返回数组。

```python
def img2vector(filename):
    returnVect = zeros((1, 1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0, 32*i+j] = int(lineStr[j])
    return returnVect
```

### 测试算法：使用k-NN识别手写数字

```python
#测试分类器
def handwritingClassTest():
    hwLabels = []
    #获取目录内容
    trainingFileList = listdir('code/trainingDigits')
    #打乱训练集数据
    random.shuffle(trainingFileList)
    print(trainingFileList)
    m = len(trainingFileList)
    trainingMat = zeros((m, 1024))
    
    for i in range(m):
        #从文件解析分类数字
        fileNameStr = trainingFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        hwLabels.append(classNumStr)
        trainingMat[i,:] = img2vector('code/trainingDigits/%s' % fileNameStr)
    
    testFileList = listdir('code/testDigits')
    errorCount = 0.0
    mTest = len(testFileList)
    #print(hwLabels)
    for i in range(mTest):
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest = img2vector('code/testDigits/%s' % fileNameStr)
        classifierResult = classify0(vectorUnderTest, trainingMat[0:1500], hwLabels, 3)
        print("the classifier came back with: %d, the real answer is: %d"
                 % (classifierResult, classNumStr))
        if(classifierResult != classNumStr): errorCount += 1.0
    print("\n 错误分类总数为: %d" % errorCount)
    print("\n 错误率为: %f" % (errorCount/float(mTest)))
```

```python
handwritingClassTest()
```
